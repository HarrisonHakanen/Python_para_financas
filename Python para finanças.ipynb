```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import yfinance as yf
```

```python
gol_df = data.DataReader(name = 'HNT-USD', data_source='yahoo', start='2015-01-01')
```

```python
hnt_df = yf.download("HNT-USD", start='2015-01-01')
hnt_df.to_csv('hnt.csv')
```

```python
moedas = ["HOT1-USD","HNT-USD","XEC-USD","SHIB-USD","NEXO-USD","SOL-USD"]

moedas_df = pd.DataFrame()
for moeda in moedas:
    moedas_df[moeda] = yf.download(moeda, start='2015-01-01')['Close']
```

```python
moedas_df
```

```python
moedas_df = moedas_df.rename(columns={'HOT1-USD':'HOLO','HNT-USD':'HELIUM','XEC-USD':'eCash','SHIB-USD':'SHIBA','NEXO-USD':'NEXO','SOL-USD':'SOLANA'})
```

```python
moedas_df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10,50))

i = 1
for i in np.arange(1,len(moedas_df.columns)):
    plt.subplot(7,1,i+1)
    sns.histplot(moedas_df[moedas_df.columns[i]])
    plt.title(moedas_df.columns[i])
```

```python
moedas_df2 = pd.read_csv('moedas.csv')
```

```python
moedas_df2.plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das moedas');
```

```python
moedas_df_normalizado = moedas_df2.copy()
for i in moedas_df_normalizado.columns[1:]:
    moedas_df_normalizado[i] = moedas_df_normalizado[i] / moedas_df_normalizado[i][0]
```

```python
moedas_df_normalizado
```

```python
moedas_df_normalizado.plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das moedas - normalizado');
```
